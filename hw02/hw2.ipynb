```python
cmap = plt.get_cmap("tab10").colors

plt.figure(figsize=(15, 10))

#  
lam = 21

#  n   (  n   )
for i, n in enumerate([30, 50, 100, 500, 1000, 5000]):
    
    p = lam / n
    
#           0  30  n   
    binom = [st.binom.pmf(k, p=p, n=n) for k in range(30)] 
    poiss = [st.poisson.pmf(k, lam) for k in range(30)] 
    
    plt.subplot(2, 3, i+1)
    plt.tight_layout()
    
#       
    plt.plot(list(range(30)), poiss, '.-', c='k', label="Poisson")
    plt.scatter(list(range(30)), binom, marker='s', color=cmap[i], label="Bin, n = " + str(n))
    
    plt.xlabel("k")
    plt.ylabel(r"$P(\xi_n=k)$")
    plt.legend()
plt.show()
```
